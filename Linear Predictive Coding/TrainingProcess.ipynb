```python
# import packages
import torch
import numpy as np
import torchvision
import matplotlib.pyplot as plt
import time
import types
from numba import jit
import matplotlib as mpl
```

```python
#import MNIST data
MNIST = torchvision.datasets.MNIST('./data/', train = True, download = True, transform = torchvision.transforms.ToTensor())   
x = [i[0].reshape(784) for i in MNIST]
x = torch.stack(x)
```

```python
# Experiments of eta = 1
N = 784
y = x
Cy = (y).t().mm(y).double().cuda()
ld,u = torch.linalg.eigh(Cy)
ld /= x.shape[0]
originalnorm = (ld.reshape(1,-1)*u**2).mean()

eta = 1
torch.manual_seed(0)
w = torch.zeros([784,784]).cuda().double()*(1 - torch.eye(784)).cuda()
v = torch.linalg.inv(torch.eye(784).cuda().double() + w)
z = v.mm(u)

t1 = time.time( )
lr = 1e-3
err_list = []
w_list = []
np.random.seed(0)

interval = 1000
n_r = []
epoch = 0
while(epoch<250000):
    epoch  += 1
    grad = -(z.mm(torch.diag(ld)).mm(z.t()).mm(v)).t()
    w -= (grad +eta/N*w)*(1-torch.eye(784).cuda())*lr   
    
    if epoch % interval == 0:
        t2 = time.time()
        min_eigv = torch.linalg.eig(torch.eye(784).cuda()*1 + w)[0].real.min()
        if min_eigv > 0:
            print(epoch, ((z**2).mean(0)*ld).mean()/originalnorm, t2 - t1, w.max(), abs((grad +eta/N*w)*(1 - torch.eye(784).cuda())).max(), abs(w).mean(), min_eigv)
            t1=t2
            err_list.append(((z**2).mean(0)*ld).mean())
            w_list.append(w.cpu().numpy()*1)
            n_r.append(epoch)
        else:
            if lr<1e-7:
                break
            lr/=2
            interval = 100
            w = torch.from_numpy(w_list[-1]).cuda()
            epoch = n_r[-1]*1
            del err_list[-1], w_list[-1], n_r[-1]
    v = torch.linalg.inv(torch.eye(784).cuda().double()*1 + w)
    z = v.mm(u)  
            
torch.save([w, w_list, err_list, n_r], 't4_11_regular='+str(eta))
```
